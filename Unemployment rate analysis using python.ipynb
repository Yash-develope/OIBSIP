```python
file41 = file1.groupby(['month'])[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']].mean()
file41 = pd.DataFrame(file41).reset_index()

month = file41.month
unemployment_rate = file41['Estimated Unemployment Rate']
labor_participation_rate = file41['Estimated Labour Participation Rate']

fig = go.Figure()

fig.add_trace(go.Bar(x=month, y=unemployment_rate, name='Unemployment Rate'))
fig.add_trace(go.Bar(x=month, y=labor_participation_rate, name='Labor Participation Rate'))

fig.update_layout(title='Unemployment Rate and Labor Participation Rate', xaxis={'categoryorder': 'array', 'categoryarray': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']})
fig.show()
```
